```python
#Time Series Analysis
# Convert date columns to datetime
covid_df['Start Date'] = pd.to_datetime(covid_df['Start Date'],
errors='coerce')
covid_df['Primary Completion Date'] = pd.to_datetime(covid_df['Primary Completion Date'], errors='coerce')

```

```python
# Plot the number of trials started over time
covid_df['Start Date'].dt.to_period('M').value_counts().sort_index().plot(kind= 'line', title='Trials Started Over Time')
```

```python

#start dates, primary completion dates, and overall completion dates analysis
covid_df['Start Date'] = pd.to_datetime(covid_df['Start Date'], errors='coerce')
covid_df['Primary Completion Date'] = pd.to_datetime(covid_df['Primary Completion Date'], errors='coerce')
covid_df['Completion Date'] = pd.to_datetime(covid_df['Completion Date'], errors='coerce')

```

```python
covid_df['Start Year-Month'] = covid_df['Start Date'].dt.to_period('M')
covid_df['Primary Completion Year-Month'] = covid_df['Primary Completion Date'].dt.to_period('M')
covid_df['Completion Year-Month'] = covid_df['Completion Date'].dt.to_period('M')
```

```python
trials_started = covid_df['Start Year-Month'].value_counts().sort_index()
trials_primary_completed = covid_df['Primary Completion Year-Month'].value_counts().sort_index()
trials_completed = covid_df['Completion Year-Month'].value_counts().sort_index()
```

```python
plt.figure(figsize=(14, 8))
plt.plot(trials_started.index.astype(str), trials_started.values, label='Trials Started', marker='o')
plt.plot(trials_primary_completed.index.astype(str), trials_primary_completed.values, label='Primary Completion', marker='o')
plt.plot(trials_completed.index.astype(str), trials_completed.values, label='Completion', marker='o')
plt.xlabel('Year-Month')
plt.ylabel('Number of Trials')
plt.title('Trends in Trial Start, Primary Completion, and Completion Dates')

plt.legend()
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.locator_params(axis='x', nbins=10)
plt.tight_layout()

plt.show()
```
